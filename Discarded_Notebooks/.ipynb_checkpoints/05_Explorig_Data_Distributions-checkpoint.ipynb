```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../csv/semi_modeled_descriptor_dataset.csv')
df
```

Now we have our modeled dataset, let's play around with the data distributions a bit and see if we can get a sense of how could it behave in a high-dimensional space. Let's observe how it behaves as we use our continuous and binary features. By "binary", we mean features that present either value X or Y (rhythmic, apathetic, poetic...).

### 1. 2D PLOT (2 continuous features)
- 'energetic/raw/rebellious/noisy/angry-warm/calm/natural/acoustic/pastoral'
- 'cold/dark/sad/atmospheric/anxious/mysterious/serious-quirky/happy'

```python
# Getting column names
X1 = 'energetic/raw/rebellious/noisy/angry-warm/calm/natural/acoustic/pastoral'
X2 = 'cold/dark/sad/atmospheric/anxious/mysterious/serious-quirky/happy'

# Extracting coordinates
x_coords = df[X1]
y_coords = df[X2]

# Creating the plot
plt.figure(figsize=(8, 6))
sns.histplot(x=x_coords, y=y_coords, bins=50, cmap='viridis', cbar=True)
plt.title("Histograma 2D")
plt.xlabel(X1)
plt.ylabel(X2)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(x_coords, y_coords, c='blue', alpha=0.6)  # Scatterplot
plt.title("Scatterplot 2D")
plt.xlabel(X1)
plt.ylabel(X2)
plt.grid(True)
plt.show()
```

### 2. 3D PLOT (2 continuous features and 1 binary feature)
- 'energetic/raw/rebellious/noisy/angry-warm/calm/natural/acoustic/pastoral'
- 'cold/dark/sad/atmospheric/anxious/mysterious/serious-quirky/happy'
- 'rhythmic'

```python
from mpl_toolkits.mplot3d import Axes3D

# Getting column names
X1 = 'energetic/raw/rebellious/noisy/angry-warm/calm/natural/acoustic/pastoral'
X2 = 'cold/dark/sad/atmospheric/anxious/mysterious/serious-quirky/happy'
X3 = 'rhythmic'

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

x_coords = df[X1]
y_coords = df[X2]
z_coords = df[X3]

ax.scatter(x_coords, y_coords, z_coords, c='blue', alpha=0.6)

ax.set_title("Scatterplot 3D")
ax.set_xlabel(X1)
ax.set_ylabel(X2)
ax.set_zlabel(X3)

plt.show()
```

Notice that this 3D plot doesn't have the "rhythmic" axis in the same visual scale as the other 2, so albums that present the "rhythmic" feature appear more distant to the rest than they actually are.

### 3. 2D Plot (2 binary features)
- 'rhythmic'
- 'poetic'

```python
# Getting column names
X1 = 'rhythmic'
X2 = 'poetic'

# Extracting coordinates
x_coords = df[X1]
y_coords = df[X2]

# Creating the plot
plt.figure(figsize=(8, 6))
sns.histplot(x=x_coords, y=y_coords, bins=50, cmap='viridis', cbar=True)
plt.title("Histograma 2D")
plt.xlabel(X1)
plt.ylabel(X2)
plt.grid(True)
plt.show()
```

### 4. 3D Plot (3 binary features)
- 'rhythmic'
- 'poetic'
- 'eclectic'

```python
# Getting column names
X1 = 'rhythmic'
X2 = 'poetic'
X3 = 'eclectic'

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

x_coords = df[X1]
y_coords = df[X2]
z_coords = df[X3]

ax.scatter(x_coords, y_coords, z_coords, c='blue', alpha=0.6)

ax.set_title("Scatterplot 3D")
ax.set_xlabel(X1)
ax.set_ylabel(X2)
ax.set_zlabel(X3)

plt.show()
```

Looking at how binary features behave on the vector space, we can observe that data sits only at specific focal points, thanks to the nature of binary data. That may seem like they give us the most defined clusters, however binary data presents a challenge for clustering algorithms that's hard to overcome, both for centroid-based clustering (https://www.ibm.com/support/pages/clustering-binary-data-k-means-should-be-avoided) and density-based clustering (https://stats.stackexchange.com/questions/330093/is-a-multicollinearity-and-or-b-binary-variables-an-issue-for-dbscan-if-so). For this reason, we won't be using any of the binary features.

Another fact that's worth noting is, as can be observed in the 2D histogram, data distribution among focal points isn't even: some focal points have a higher concentration on data points than others. This should be anticipated, since the histograms for single binary features already showed us that distribution there wasn't even, so when combining different binary features, the same effect will happen, just in higher dimensions.
